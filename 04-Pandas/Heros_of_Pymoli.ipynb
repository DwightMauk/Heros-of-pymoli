```python
import pandas as pd
import numpy as np

csv_file = "../Resources/purchase_data.csv"

heros_data = pd.read_csv(csv_file)

heros_data.head()
```

```python
#player count
heros_data['SN'].value_counts()
#576 users
player_count = len(heros_data['SN'].value_counts())
total_players = pd.DataFrame({'Total Players':[player_count]})
total_players
```

```python
heros_data['Item ID'].unique()
#183 items
```

```python
heros_data['Price'].mean()
#avg price 3.05
```

```python
heros_data['Price'].sum()
#total revenue 2379.77
```

```python
heros_data['Item ID'].count()
# 780 item sold
```

```python
items = len((heros_data['Item ID']).unique())
avg_price = (heros_data['Price']).mean()
total_revenue = (heros_data['Price']).sum()
item_sold = (heros_data['Purchase ID']).count()

summary_df = pd.DataFrame({'Number of Unique Items':[items],'Average Price':[avg_price], 'Number of Purchases': [item_sold], 
                           'Total Revenue': [total_revenue]})

summary_df['Average Price'] = summary_df['Average Price'].astype(float).map("${:,.2f}".format)
summary_df['Total Revenue'] = summary_df['Total Revenue'].astype(float).map("${:,.2f}".format)

summary_df
```

```python
heros_data['Gender'].value_counts()
```

```python
gender_data = heros_data.groupby('Gender')

gender_count = gender_data.nunique()["SN"]

gender_summary = pd.DataFrame({'Total Count':gender_count,
                               'Percentage of players':gender_count/player_count*100})


gender_summary['Total Count'] = gender_summary.sort_values(['Total Count'], ascending = False)
gender_summary['Percentage of players'] = gender_summary['Percentage of players'].astype(float).map("{:,.2f}%".format)

gender_summary
```

```python
item_gender = gender_data['Purchase ID'].count()

avg_price_gender = gender_data['Price'].mean()

total_gender = gender_data['Price'].sum()

avg_per_gender = total_gender/gender_count

gender_sum_2 = pd.DataFrame({'Purchase Count': item_gender, 'Average Purchase Price': avg_price_gender,
                                    'Total Purchase Value':total_gender,'Avg Purchase Total per Person': avg_per_gender})

gender_sum_2.index.name = 'Gender'

gender_sum_2['Average Purchase Price'] = gender_sum_2['Average Purchase Price'].astype(float).map("${:,.2f}".format)
gender_sum_2['Total Purchase Value'] = gender_sum_2['Total Purchase Value'].astype(float).map("${:,.2f}".format)
gender_sum_2['Avg Purchase Total per Person'] = gender_sum_2['Avg Purchase Total per Person'].astype(float).map("${:,.2f}".format)

gender_sum_2
```

```python
age_df = pd.DataFrame(heros_data)
bins = [0,10,14,19,24,29,34,39,100]
age_group = ['<10','11-14','15-19','20-24','25-29','30-34','35-39','40+']
pd.cut(heros_data['Age'], bins, labels=age_group).head()
```

```python
heros_data['Age Group'] = pd.cut(heros_data['Age'], bins, labels=age_group)
heros_data.head()
```

```python
group_count = heros_data.groupby('Age Group')

total_count_age = group_count['SN'].nunique()

percentage_by_age = (total_count_age/player_count) * 100

age_summary = pd.DataFrame({'Total Count': total_count_age,'Percentage of Players': percentage_by_age})

age_summary['Percentage of Players'] =  age_summary['Percentage of Players'].astype(float).map("{:,.2f}%".format)

age_summary
```

```python
item_age = group_count['Purchase ID'].count()

avg_price_age = group_count['Price'].mean()

total_price_age = group_count['Price'].sum()

avg_per_age = total_price_age/total_count_age

age_sum_2 = pd.DataFrame({'Purchase Count': item_age, 'Average Purchase Price': avg_price_age,
                                    'Total Purchase Value':total_price_age,'Avg Purchase Total per Person': avg_per_age})

age_sum_2.index.name = 'Age'

age_sum_2['Average Purchase Price'] = age_sum_2['Average Purchase Price'].astype(float).map("${:,.2f}".format)
age_sum_2['Total Purchase Value'] = age_sum_2['Total Purchase Value'].astype(float).map("${:,.2f}".format)
age_sum_2['Avg Purchase Total per Person'] = age_sum_2['Avg Purchase Total per Person'].astype(float).map("${:,.2f}".format)

age_sum_2
```

```python
top_spender = heros_data.groupby('SN')

count_of_item = top_spender['Purchase ID'].count()

avg_purchase = top_spender['Price'].mean()

total_purchase = top_spender['Price'].sum()

top_spender_data = pd.DataFrame({'Purchase Count': count_of_item,'Average Purchase Price': avg_purchase,
                                 'Total Purchase Value':total_purchase})

format_spender = top_spender_data.sort_values(["Total Purchase Value"], ascending=False).head()

format_spender['Average Purchase Price'] = format_spender['Average Purchase Price'].astype(float).map("${:,.2f}".format)
format_spender['Total Purchase Value'] = format_spender['Total Purchase Value'].astype(float).map("${:,.2f}".format)

format_spender
```

```python
items = heros_data[['Item ID', 'Item Name', 'Price']]

top_item = heros_data.groupby(['Item Name','Item ID'])

count_top_item = top_item['Price'].count()

purchase_item = (top_item['Price'].sum())

item_price = purchase_item/count_top_item

top_item_data = pd.DataFrame({'Purchase Count': count_top_item, 'Item Price': item_price,'Total Purchase Value':purchase_item})

format_item = top_item_data.sort_values(["Purchase Count"], ascending=False).head()

format_item['Total Purchase Value'] = format_item['Total Purchase Value'].astype(float).map("${:,.2f}".format)

format_item
```
